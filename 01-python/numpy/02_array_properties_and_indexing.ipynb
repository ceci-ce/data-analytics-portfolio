# 1. Array Properties

NumPy arrays contain useful attributes that allow us to understand their structure and characteristics.

These properties are essential when working with datasets because they help us identify the dimensions, size and data types before performing any analysis.

````python
import numpy as np

# Create a sales dataset

sales = np.array([
    [120, 135, 150, 160],
    [98, 110, 125, 140],
    [200, 220, 240, 260]
])

print(sales)
````

## Shape

The `shape` attribute returns the dimensions of an array.

It tells us the number of rows and columns in a two-dimensional array.

The result is returned as a tuple.

````python
sales.shape
````

The array contains:

- 3 rows
- 4 columns

Therefore, the total structure is 3 × 4.

## Size

The `size` attribute returns the total number of elements contained in an array.

````python
sales.size
````

The array contains 12 values because:

3 rows × 4 columns = 12 elements

## Number of dimensions (`ndim`)

The `ndim` attribute returns the number of dimensions of an array.

Examples:

- 1 dimension → vector
- 2 dimensions → matrix
- 3 dimensions → multidimensional array

````python
sales.ndim
````

````python
one_dimension = np.array([10, 20, 30])

three_dimensions = np.array([
    [
        [1, 2],
        [3, 4]
    ]
])

print(one_dimension.ndim)

print(three_dimensions.ndim)
````

## Data type (`dtype`)

The `dtype` attribute shows the data type stored inside the array.

Common NumPy data types:

- `int` → integers
- `float` → decimal numbers
- `bool` → True/False values
- `str` → text values

````python
sales.dtype
````

````python
prices = np.array([
    19.99,
    25.50,
    32.75
])

prices.dtype
````

## Memory usage

NumPy arrays are efficient because they store data in a compact way.

- `itemsize` returns the size in bytes of each element.
- `nbytes` returns the total memory used by the array.

# 2. Indexing

Indexing allows us to access individual elements from a NumPy array.

The indexing system in NumPy is similar to Python lists:

- Indexing starts at 0.
- Negative indexes can be used to access elements from the end.

````python
import numpy as np

sales = np.array([120, 135, 150, 160, 180])

print(sales)
````

### 1D Array Indexing

````python
sales[0]
````

````python
sales[-1]
````

In NumPy:

- `array[0]` → first element
- `array[-1]` → last element

### 2D Array Indexing

````python
monthly_sales = np.array([
    [120, 135, 150],
    [200, 220, 250],
    [90, 110, 130]
])

print(monthly_sales)
````

````python
monthly_sales[1,2]
````

The first index represents the row.

The second index represents the column.

`array[row, column]`

````python
monthly_sales[0]
````

````python
monthly_sales[:,1]
````

The colon `:` means "select all values".

Example:

`array[:,1]`

means:

- all rows
- column with index 1

### 3D Array Indexing

````python
sales_3d = np.array([
    [
        [120,130],
        [140,150]
    ],
    [
        [200,220],
        [240,260]
    ]
])

print(sales_3d)
````

````python
sales_3d[0]
````

````python
sales_3d[1,0,1]
````

For 3D arrays:

array[dimension, row, column]


# 3. Slicing

Similar to Python lists, NumPy arrays allow us to access individual elements using square brackets `[]`.

NumPy also supports **slicing**, which allows us to extract portions of an array and create subarrays.

The general slicing syntax is:

```python
array[start:stop:step]
```

Where:

- `start` → the index where the slice begins (included).
- `stop` → the index where the slice ends (excluded).
- `step` → the interval between selected elements.

If any of these values are not specified, NumPy uses the default values:

- `start = 0`
- `stop = size of the dimension`
- `step = 1`

Slicing works with one-dimensional arrays as well as multi-dimensional arrays, allowing us to select specific rows, columns, or sections of a dataset.

````python
sales = np.array([100,200,300,400,500])

sales[1:4]
````

## Solutions

````python
# Solution 1

sales = np.array([250, 300, 280, 350, 400])

first_value = sales[0]
last_value = sales[-1]
third_value = sales[2]

print(f"First value: {first_value}")
print(f"Last value: {last_value}")
print(f"Third value: {third_value}")
````

````python
# Solution 2

stores_sales = np.array([
    [120, 135, 150],
    [200, 220, 250],
    [90, 110, 130]
])

# First store sales (first row)
first_store = stores_sales[0]

# Second month sales (second column)
second_month = stores_sales[:,1]

# Last column
last_column = stores_sales[:,-1]

print(f"First store sales: {first_store}")
print(f"\nSecond month sales: {second_month}")
print(f"\nLast column: {last_column}")
````

````python
# Solution 3

sales_3d = np.array([
    [
        [120, 130, 140, 150],
        [80, 90, 100, 110],
        [200, 220, 240, 260]
    ],
    [
        [150, 160, 170, 180],
        [95, 105, 115, 125],
        [250, 270, 290, 310]
    ]
])


store_2_product_1_week_3 = sales_3d[1,0,2]


print(store_2_product_1_week_3)
````

````python
# Challenge Solution

customer_ages = np.array([
    22,
    35,
    28,
    41,
    30,
    26
])

# First three customers
first_three = customer_ages[:3]


# Last two customers
last_two = customer_ages[-2:]


# Every second customer
every_second = customer_ages[::2]


print("First three:")
print(first_three)

print("\nLast two:")
print(last_two)

print("\nEvery second customer:")
print(every_second)
````

## Key Takeaways

NumPy indexing allows us to access and manipulate specific parts of an array.

Main concepts:

- `array[index]` → access a single element
- `array[row, column]` → access elements in 2D arrays
- `:` → select all values in a dimension
- Negative indexes → access elements from the end
- Slicing → extract subsets of data

These concepts are essential for data manipulation and are the foundation for working with Pandas DataFrames.
