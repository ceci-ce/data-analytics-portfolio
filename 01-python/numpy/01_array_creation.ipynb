# 3. Create arrays from Python lists


One of the easiest ways to create a NumPy array is by converting a standard Python list.

NumPy arrays are more memory-efficient and support fast mathematical operations, making them the preferred data structure for data analysis.

```python
import numpy as np

# Daily website visitors
daily_visitors = [245, 312, 298, 401, 389, 420, 375]

visitors_array = np.array(daily_visitors)

print(visitors_array)
```

The result is a one-dimensional NumPy array.

```python
# Monthly revenue (thousands of €)

monthly_revenue = [18.5, 20.1, 21.8, 19.9]

revenue_array = np.array(monthly_revenue)

print(revenue_array)
```

NumPy automatically detects the appropriate data type.

```python
# Sales of three stores during four months

sales = [
    [120, 135, 142, 150],
    [98, 110, 115, 123],
    [160, 158, 171, 180]
]

sales_array = np.array(sales)

print(sales_array)
```

This is a two-dimensional array (matrix).

```python
# Sales by region, store and quarter

sales_3d = np.array([
    [
        [120, 130],
        [140, 150]
    ],
    [
        [160, 170],
        [180, 190]
    ]
])

print(sales_3d)
```

### Summary

NumPy arrays can be created directly from Python lists.

Arrays can have:

- One dimension (vector)
- Two dimensions (matrix)
- Three or more dimensions

# 4. Create arrays with NumPy functions

NumPy provides several built-in functions to create arrays quickly without manually typing every value.

These functions are especially useful when working with large datasets, initializing matrices or generating sequences of numbers.


```python
import numpy as np

# Create a 3x4 matrix filled with zeros

zeros_matrix = np.zeros((3, 4))

print(zeros_matrix)
```

`np.zeros()` creates an array filled with zeros.

```python
# Create a 2x5 matrix filled with ones

ones_matrix = np.ones((2, 5))

print(ones_matrix)
```

`np.ones()` creates an array where every value is equal to one.

```python
# Create a 3x3 matrix filled with the value 7

full_matrix = np.full((3, 3), 7)

print(full_matrix)
```

`np.full()` fills the array with any value you choose.

```python
import numpy as np

# Create a 4x4 identity matrix

identity_matrix = np.eye(4)

print(identity_matrix)
```

Identity matrices are commonly used in linear algebra.

```python
# Generate even numbers from 0 to 20

even_numbers = np.arange(0, 21, 2)

print(even_numbers)
```

`np.arange(start, stop, step)` creates an array of evenly spaced values.

In this example:

- `0` is the starting value.
- `21` is the stopping value (**not included**).
- `2` is the step, so only even numbers are generated.

> **Note:** Just like Python's `range()` function, the `stop` value is **excluded**. That's why `21` is not part of the resulting array, and the last value is `20`.

```python
# Generate six equally spaced numbers between 0 and 100

values = np.linspace(0, 100, 6)

print(values)
```

###  `np.arange()` vs `np.linspace()`

Although both functions generate sequences of numbers, they are used in different situations.

| Function | Best used when... |
|----------|-------------------|
| `np.arange()` | You know the **step size** between values. |
| `np.linspace()` | You know the **exact number of values** you want to generate. |

**Example:**

- `np.arange(0, 10, 2)` → Creates numbers every **2 units**.
- `np.linspace(0, 10, 6)` → Creates **6 equally spaced values** between 0 and 10.

# 5. Random arrays

NumPy provides several functions to generate random values.

Random arrays are commonly used to simulate data, test algorithms and create sample datasets for analysis.

```python
import numpy as np

# Set a random seed for reproducibility

np.random.seed(42)
```

Using a random seed ensures that the same random numbers are generated every time the code is executed.

```python
# Generate five random numbers between 0 and 1

random_values = np.random.rand(5)

print(random_values)
```

```python
# Generate ten random customer IDs between 1000 and 9999

customer_ids = np.random.randint(1000, 10000, size=10)

print(customer_ids)
```

```python
# Generate a 3x4 matrix with random integers

sales = np.random.randint(100, 500, size=(3, 4))

print(sales)
```

```python
# Randomly select products

products = np.array([
    "Laptop",
    "Tablet",
    "Smartphone",
    "Monitor",
    "Keyboard"
])

selected_products = np.random.choice(products, size=3)

print(selected_products)
```
