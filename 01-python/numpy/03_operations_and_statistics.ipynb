# 1. Array Operations

NumPy allows us to perform mathematical operations directly on arrays.

These operations are vectorized, meaning that calculations are applied to all elements at once without the need for explicit loops.

This makes NumPy much faster and more efficient than traditional Python lists.

### Import Numpy and create arrays

```python
import numpy as np

sales_january = np.array([120, 150, 170, 200])

sales_february = np.array([130, 160, 180, 220])

sales_january
```

## Addition and subtraction

NumPy performs operations element by element.

When two arrays have the same shape, the first element is operated with the first element, the second with the second, and so on.

```python
# Addition

total_sales = sales_january + sales_february

print(total_sales)
```

```python
# Subtraction

difference = sales_february - sales_january

print(difference)
```

## Multiplication and division

Multiplication and division are also performed element by element.

```python
# multiplication

products = np.array([10, 20, 30])

prices = np.array([5, 8, 12])


revenue = products * prices

print(revenue)
```

```python
# division

average_price = revenue / products

print(average_price)
```

## Operations with scalars

A scalar is a single value.

NumPy automatically applies the operation to every element in the array.

```python
sales = np.array([100, 200, 300])

sales * 2
```

```python
discount = sales * 0.9

discount
```

This operation applies a 10% discount to every value in the array.

## Vectorized operations

Vectorization allows us to perform calculations on entire arrays instead of using loops.

Example:

Traditional Python:

```python
sales = [100, 200, 300]

new_sales = []

for value in sales:
    new_sales.append(value * 2)

print(new_sales)
```

With NumPy:

```python
sales_array = np.array([100, 200, 300])

sales_array * 2
```

## Mathematical functions

NumPy provides many mathematical functions that can be applied directly to arrays.

### Square root

```python
numbers = np.array([1, 4, 9, 16])

np.sqrt(numbers)
```

### Exponential

```python
np.exp(numbers)
```

### Absolute values

```python
values = np.array([-5, -10, 15])

np.abs(values)
```

### Power

```python
np.power(numbers, 2)
```

# 2. Statistical Operations

NumPy provides several functions to perform basic statistical analysis on numerical data.

These functions are commonly used during Exploratory Data Analysis (EDA) to understand datasets and identify patterns.

### Sum

```python
sales = np.array([120, 150, 170, 200, 180])

total_sales = np.sum(sales)

print(total_sales)
```

`np.sum()` calculates the total value of all elements in an array.

### Mean (Average)

```python
average_sales = np.mean(sales)

print(average_sales)
```

The mean represents the average value of a dataset.

It is calculated by dividing the total sum by the number of elements.

### Median

```python
median_sales = np.median(sales)

print(median_sales)
```

The median represents the middle value of a dataset when the values are sorted.

Unlike the mean, the median is less affected by extreme values (outliers).

### Minimum and Maximum

```python
minimum = np.min(sales)

maximum = np.max(sales)

print(minimum)
print(maximum)
```

### Standard Deviation

```python
std_sales = np.std(sales)

std_sales
```

Standard deviation measures how spread out the values are around the mean.

A low standard deviation means that values are close to the average.

A high standard deviation indicates greater variability.

### Variance

```python
variance_sales = np.var(sales)

print(variance_sales)
```

Variance measures the average squared difference between each value and the mean.

It is related to standard deviation:

Standard deviation = √variance

### Percentiles

```python
prices = np.array([10, 20, 30, 40, 50, 60, 70])


np.percentile(prices, 50)
```

Percentiles indicate the value below which a given percentage of observations fall.

Examples:

- 50th percentile → median
- 25th percentile → first quartile
- 75th percentile → third quartile

### Correlation

```python
advertising = np.array([10,20,30,40,50])

sales = np.array([100,150,200,250,300])


correlation = np.corrcoef(advertising, sales)

print(correlation)
```

Correlation measures the relationship between two variables.

Values close to:

- 1 → strong positive relationship
- 0 → no linear relationship
- -1 → strong negative relationship

## Solutions

```python
# Solution 1

import numpy as np

daily_sales = np.array([120, 135, 150, 160, 125, 140, 130])

total_sales = np.sum(daily_sales)
average_sales = np.mean(daily_sales)
minimum_sales = np.min(daily_sales)
maximum_sales = np.max(daily_sales)

print("Total sales:", total_sales)
print("Average sales:", average_sales)
print("Minimum sales:", minimum_sales)
print("Maximum sales:", maximum_sales)
```

```python
# Solution 2

values = np.array([20, 25, 30, 35, 40])

standard_deviation = np.std(values)
variance = np.var(values)

print("Standard deviation:", standard_deviation)
print("Variance:", variance)
```

```python
# Solution 3

website_visits = np.array([1200, 1500, 1800, 2000, 2500])

mean_visits = np.mean(website_visits)
median_visits = np.median(website_visits)
percentile_75 = np.percentile(website_visits, 75)

print("Mean:", mean_visits)
print("Median:", median_visits)
print("75th percentile:", percentile_75)
```
