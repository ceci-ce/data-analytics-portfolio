# NumPy Business Case: Retail Sales Performance Analysis

## Business Scenario

A retail company wants to analyze its sales performance during a 30-day period.

The objective is to understand:

- Total revenue
- Average daily sales
- Best and worst performing days
- Sales variability
- Days with above-average performance

NumPy will be used to generate and analyze the dataset.

## Dataset Generation

```python
import numpy as np

np.random.seed(42)

days = np.arange(1,31)

daily_sales = np.random.randint(
    1000,
    5000,
    size=30
)

customers = np.random.randint(
    50,
    200,
    size=30
)

average_ticket = daily_sales / customers
```

```python
daily_sales[:10]
```

The generated dataset represents:

- 30 business days
- Daily revenue
- Number of customers
- Average customer spending

## Data Exploration

### First values

```python
print(daily_sales[:10])
```

### Dataset dimensions

```python
print("Number of days:", daily_sales.size)
print("Dimensions:", daily_sales.ndim)
```

## Descriptive Statistics

### Total revenue

```python
total_revenue = np.sum(daily_sales)

print(total_revenue)
```

### Average daily sales

```python
average_sales = np.mean(daily_sales)

print(average_sales)
```

### Median

```python
median_sales = np.median(daily_sales)

print(median_sales)
```

### Minimum and maximum

```python
minimum_sales = np.min(daily_sales)

maximum_sales = np.max(daily_sales)


print(minimum_sales)
print(maximum_sales)
```

## Standard deviation

```python
sales_std = np.std(daily_sales)

print(sales_std)
```

Standard deviation helps us understand sales variability.

A high value means sales fluctuate significantly between days.

## Business KPIs

### Best sales day

```python
best_day = np.argmax(daily_sales)


print(
    "Best day:",
    days[best_day]
)

print(
    "Revenue:",
    daily_sales[best_day]
)
```

### Worst sales day

```python
worst_day = np.argmin(daily_sales)


print(
    "Worst day:",
    days[worst_day]
)

print(
    "Revenue:",
    daily_sales[worst_day]
)
```

### Average customers

```python
average_customers = np.mean(customers)

print(average_customers)
```

### Average ticket

```python
average_ticket_value = np.mean(average_ticket)

print(average_ticket_value)
```

## Business Questions

### Which days performed above average?

```python
above_average_days = days[
    daily_sales > average_sales
]


print(above_average_days)
```

### What percentage of days exceeded average sales?

```python
percentage = (
    np.sum(daily_sales > average_sales)
    /
    len(daily_sales)
) * 100


print(percentage)
```

### What was the revenue difference between the best and worst day?

```python
difference = (
    maximum_sales -
    minimum_sales
)

print(difference)
```

## Insights and Conclusions

```python
business_summary = {
    "Total revenue": total_revenue,
    "Average daily revenue": average_sales,
    "Best day revenue": maximum_sales,
    "Worst day revenue": minimum_sales,
    "Above average sales days (%)": percentage
}

print(business_summary)
```

### Business Insights

After analyzing the simulated retail dataset:

- The company generated a total revenue of **91,462€**.
- The average daily revenue was **3048.73€**.
- The best performing day generated **4772€**.
- Sales variability shows how much daily revenue fluctuates during the period.
- **50% of the days exceeded the average sales level.**


These insights could help business teams identify high-performing periods and support decision-making.
