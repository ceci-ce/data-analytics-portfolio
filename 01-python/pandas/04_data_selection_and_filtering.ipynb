# Data Selection and Filtering

After loading a dataset, the next step is selecting the information that is relevant for the analysis.

Pandas provides powerful tools to:

- Select specific columns
- Select rows
- Filter data based on conditions
- Sort values
- Find unique values

These operations are among the most frequently used in real-world Data Analytics projects.

````python
import pandas as pd
````

````python
customers = pd.read_csv("../../datasets/customers.csv")

customers.head()
````

# Selecting Columns

Columns can be selected using square brackets.

Selecting one column returns a Series.

Selecting multiple columns returns a DataFrame.

### Single column

````python
customers["Customer"]
````

### Multiple columns

# Using loc[]

## Label-based Selection

`.loc[]` selects rows and columns using labels.

Syntax:

```python
df.loc[row_label, column_label]
```

### Select one row

````python
customers.loc[0]
````

### Select multiple rows

````python
customers.loc[0:3]
````

### Select one value

````python
customers.loc[0, "Customer"]
````

### Select multiple columns

````python
customers.loc[
    :,
    [
        "Customer",
        "Country"
    ]
]
````

# Using iloc

## Position-based Selection

`.iloc[]` selects rows and columns using integer positions.

### First row

````python
customers.iloc[0]
````

### First three rows

````python
customers.iloc[0:3]
````

### Row and column

````python
customers.iloc[
    0,
    2
]
````

### Multiple rows and columns

````python
customers.iloc[
    0:4,
    1:3
]
````

### loc vs iloc

| Method | Uses |
|---------|------|
| loc | Labels |
| iloc | Integer positions |

Remember:

- `loc` → names
- `iloc` → numbers

# Boolean Filtering

Filtering allows analysts to keep only the rows that satisfy specific conditions.

This is one of the most common operations in Pandas.

### Customers older than 30

````python
customers[
    customers["Age"] > 30
]
````

### Customers from Spain

````python
customers[
    customers["Country"] == "Spain"
]
````

### Multiple filters

### AND

````python
customers[
    (customers["Country"] == "Spain") &
    (customers["Age"] > 30)
]
````

### OR

````python
customers[
    (customers["Country"] == "Spain") |
    (customers["Country"] == "France")
]
````

### NOT

````python
customers[
    customers["Country"] != "Spain"
]
````

# isin()

`isin()` checks whether values belong to a list.

It is much cleaner than chaining multiple OR conditions.

````python
customers[
    customers["Country"].isin(
        [
            "Spain",
            "France"
        ]
    )
]
````

# Sorting

### Sort by Age (Ascending)

````python
customers.sort_values("Age")
````

### Sort by Age (Descending)

````python
customers.sort_values(
    "Age",
    ascending=False
)
````

### Sort by Country and Age

````python
customers.sort_values(
    by=["Segment", "Age"],
    ascending=[True, False]
)
````

# Unique Values

### unique()

````python
customers["Country"].unique()
````

### nunique()

````python
customers["Country"].nunique()
````

### value_counts()

````python
customers["Country"].value_counts()
````

These methods help analysts understand categorical variables.

They are especially useful during Exploratory Data Analysis (EDA).
