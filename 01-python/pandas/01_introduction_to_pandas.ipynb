# Introduction to Pandas

Pandas is a Python library designed for data manipulation and analysis.

It provides powerful data structures that allow analysts to:

- Load datasets
- Clean data
- Transform information
- Perform calculations
- Prepare data for visualization

## Why Pandas is essential for Data Analysts

Pandas is widely used because it simplifies common analytical tasks:

- Handling missing values
- Filtering information
- Creating new variables
- Aggregating data
- Preparing datasets for BI tools such as Power BI

## Relationship between NumPy and Pandas

Pandas is built on top of NumPy.

NumPy provides the numerical foundation, while Pandas adds labeled data structures such as Series and DataFrames.

Example:

NumPy → arrays

Pandas → tables with rows and columns
