# Series and DataFrames in Pandas

Pandas provides two main data structures:

- Series
- DataFrame

These structures allow analysts to work with labeled and structured data efficiently.

A Series represents a single column of data, while a DataFrame represents a complete table composed of multiple Series.

Understanding these structures is essential because most real-world datasets are stored as DataFrames.

# Pandas Series

A Series is a one-dimensional labeled array.

It can store:

- Numbers
- Text
- Dates
- Other Python objects

Each value has an associated index, which allows efficient data selection.

## Creating a Series

### From a Python list

```python
import pandas as pd

sales = pd.Series(
    [100, 200, 300, 400]
)

print(sales)
```

By default, Pandas creates an automatic numerical index starting from 0.

The index identifies each value in the Series.

### Custom Index

```python
sales = pd.Series(
    [100, 200, 300, 400],
    index=[
        "January",
        "February",
        "March",
        "April"
    ]
)

print(sales)
```

A custom index can make data easier to understand.

In business analytics, labels are often more meaningful than numerical indexes.

## Accessing Series values

### Using index

```python
sales["March"]
```

### Using position

```python
sales.iloc[2]
```

There are two common ways to access data:

- Label-based selection → `.loc[]`
- Position-based selection → `.iloc[]`

## Series attributes

```python
print("Values:")
print(sales.values)


print("\nIndex:")
print(sales.index)


print("\nData type:")
print(sales.dtype)
```

## Series methods

### Basic statistics

```python
sales.sum()
```

```python
sales.mean()
```

```python
sales.max()
```

```python
sales.min()
```

Series objects include many methods for data analysis.

These methods are frequently used when calculating business metrics.

# Pandas DataFrame

A DataFrame is a two-dimensional data structure organized into rows and columns.

It is similar to a spreadsheet or a SQL table.

Each column in a DataFrame is a Pandas Series.

Example:

Customer database:

| Customer | Age | City |
|---|---|---|
| Ana | 25 | Madrid |
| Luis | 32 | Barcelona |

## Creating a DataFrame

### From dictionary

```python
data = {
    "Product": [
        "Laptop",
        "Phone",
        "Tablet"
    ],
    
    "Price": [
        1200,
        800,
        500
    ],
    
    "Stock": [
        10,
        25,
        15
    ]
}


df = pd.DataFrame(data)


print(df)
```

## DataFrame structure

```python
df.shape
```

`shape` returns:

(rows, columns)

In this example:

- 3 rows
- 3 columns

## Columns

### List column names

### Select a column

```python
df["Product"]
```

Selecting a single column returns a Series.

### Select multiple columns

```python
df[
    [
        "Product",
        "Price"
    ]
]
```

Selecting multiple columns returns another DataFrame.

## Adding columns

```python
df["Total Value"] = (
    df["Price"] *
    df["Stock"]
)

print(df)
```

# Business Example: Store Sales

A company stores information about its products.

We want to calculate the revenue generated by each product.

```python
sales_data = {
    "Product":[
        "Laptop",
        "Phone",
        "Tablet",
        "Monitor"
    ],
    
    "Units Sold":[
        15,
        30,
        20,
        12
    ],
    
    "Price":[
        1200,
        800,
        500,
        300
    ]
}


sales_df = pd.DataFrame(
    sales_data
)


print(sales_df)
```

```python
sales_df["Revenue"] = (
    sales_df["Units Sold"] *
    sales_df["Price"]
)

print(sales_df)
```

## Solutions

```python
# Solution 1

monthly_revenue = pd.Series(
    [5000, 7000, 6500, 8000], 
    index=[
        "January",
        "February",
        "March",
        "April"
    ]
)

print(monthly_revenue)

print(f"Total revenue: {monthly_revenue.sum():,.2f} €")
print(f"Average revenue: {monthly_revenue.mean():,.2f} €")
print(f"Best month: {monthly_revenue.idxmax()}")
print(f"Highest revenue: {monthly_revenue.max():,.2f} €")
```

```python
# Solution 2

home_decor = {
    "Product": [
        "Ceramic Vase",
        "Wall Mirror",
        "Table Lamp",
        "Cushion Cover",
        "Wooden Shelf",
        "Scented Candle"
    ],
    "Category": [
        "Decor",
        "Wall Decor",
        "Lighting",
        "Textiles",
        "Furniture",
        "Home Fragrance"
    ],
    "Price": [
        35.99,
        89.50,
        59.90,
        18.99,
        74.95,
        12.50
    ],
    "Quantity": [
        25,
        10,
        18,
        40,
        12,
        50
    ]
}

decor_df = pd.DataFrame(home_decor)

decor_df["Revenue"] = (
    decor_df["Price"] *
    decor_df["Quantity"]
) 

print(decor_df)

average_revenue = decor_df["Revenue"].mean()

print(f"\nAverage revenue: {average_revenue}")

print("\nProducts above average revenue:")

print(
    decor_df[
        decor_df["Revenue"] > average_revenue
    ]
)
```

```python
# Solution 3

customers = {
    "Customer": [
        "Alice",
        "Bob",
        "Charlie",
        "Diana",
        "Emma"
    ],
    
    "Age": [
        25,
        34,
        29,
        41,
        31
    ],
    
    "Country": [
        "Spain",
        "France",
        "Spain",
        "Germany",
        "Spain"
    ],
    
    "Purchases": [
        1200,
        950,
        1800,
        700,
        1500
    ]
}

customers_df = pd.DataFrame(customers)

print(customers_df)

average_age = customers_df["Age"].mean()

print(f"\nAverage age: {average_age}")

highest_purchase = customers_df.loc[
    customers_df["Purchases"].idxmax()
    ]

print(f"\nCustomer with highest purchases:")
print(highest_purchase)

print("\nCustomers by country:")
print(customers_df["Country"].value_counts())
```
