# Loading and Exploring Data

One of the first tasks of a Data Analyst is loading and understanding a dataset.

Before performing any analysis, it is essential to explore the data to understand:

- Dataset size
- Available columns
- Data types
- Missing values
- General statistics

This initial exploration helps identify potential data quality issues and guides the next steps of the analysis.

```python
import pandas as pd
```

# Loading Data

```python
import pandas as pd

# Load the dataset
sales_df = pd.read_csv("../../datasets/sales.csv")

# Number of unique categories
print("Unique categories:")
print(sales_df["Category"].nunique())

# Products per category
print("\nProducts per category:")
print(sales_df["Category"].value_counts())

# Missing values
print("\nMissing values:")
print(sales_df.isnull().sum())

# Average product price
print("\nAverage product price:")
print(sales_df["Price"].mean())
```

```python
# Extra Challenge

import pandas as pd

# Load the dataset
sales_df = pd.read_csv("../../datasets/sales.csv")

# Create a new Revenue column
sales_df["Revenue"] = sales_df["Price"] * sales_df["Quantity"]

print(sales_df)

# Product with the highest revenue
highest_revenue = sales_df.loc[sales_df["Revenue"].idxmax()]

print("\nProduct with highest revenue:")
print(highest_revenue)

# Total revenue
print("\nTotal revenue:")
print(sales_df["Revenue"].sum())

# Average revenue
average_revenue = sales_df["Revenue"].mean()

print("\nAverage revenue:")
print(average_revenue)

# Products above average revenue
print("\nProducts above average revenue:")
print(sales_df[sales_df["Revenue"] > average_revenue])
```
